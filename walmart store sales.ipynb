```python
#features and stores have a commom key 'Stores' and store in a variable features_store
features_store = features.merge(stores, how='inner', on='Store')
```

```python
# Date field has string type. We can convert them to datetime
train.head(8)
```

```python
# check to see if 'train' and 'test' dataframes has 'Date' type to convert
pd.DataFrame({'Type_Train': train.dtypes, 'Type_Test': test.dtypes})
```

```python
features_store.Date = pd.to_datetime(features_store.Date)
train.Date = pd.to_datetime(train.Date)
test.Date = pd.to_datetime(test.Date)
```

```python
# Date field represent each week, ending every friday.
#create a 'Week'  and  'Year' field
features_store['Week'] = features_store.Date.dt.week 
features_store['Year'] = features_store.Date.dt.year
```

```python
#create two new dataframes 'train_new_details' and 'test_new_details' 
#containing all the information needed 
#join 'train' and 'test' with 'features_store'
#keys are 'Store', 'Dept' and 'IsHoliday'.
train_new_details = train.merge(features_store,
                     how='inner',
                     on=['Store','Date','IsHoliday']).sort_values(by=['Store','Dept','Date']).reset_index(drop=True)
```

```python
test_new_details = test.merge(features_store, 
                           how='inner', 
                           on=['Store','Date','IsHoliday']).sort_values(by=['Store','Dept','Date']).reset_index(drop=True)
```

```python
del features, train, stores, test
```

```python
#search for null values for each column.
null_columns = (train_new_details.isnull().sum(axis = 0)/len(train_new_details)).sort_values(ascending=False).index
null_data = pd.concat([
    train_new_details.isnull().sum(axis = 0),
    (train_new_details.isnull().sum(axis = 0)/len(train_new_details)).sort_values(ascending=False),
    train_new_details.loc[:, train_new_details.columns.isin(list(null_columns))].dtypes], axis=1)
null_data = null_data.rename(columns={0: '# null', 
                                      1: '% null', 
                                      2: 'type'}).sort_values(ascending=False, by = '% null')
null_data = null_data[null_data["# null"]!=0]
null_data
```
